```python
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
```

```python
SSH_FILTERED = '/work/ab0995/b382485/eddy_data/output_path/months/ssh_filtered_gridded_2015_001_01.nc'
SSH_RAW = '/work/ab0995/b382485/eddy_data/output_path_raw/months/ssh_filtered_gridded_2015_001_01.nc'
```

```python
raw = xr.open_dataset(SSH_RAW)
filtered = xr.open_dataset(SSH_FILTERED)
```

```python
raw.ssh.description
```

```python
plot_data = raw.ssh.values[0]

fig = plt.figure(figsize=(12, 4))
plt.imshow(np.flipud(plot_data.T), cmap=cm.seismic, vmin=-1, vmax=1)
cb = plt.colorbar(orientation='vertical', fraction=0.046, pad=0.02, label='SSH (m)')
plt.title("Time: 2015-01-01", fontsize=10)
plt.show()

plot_data = filtered.ssh.values[0]

fig = plt.figure(figsize=(12, 4))
plt.imshow(np.flipud(plot_data.T), cmap=cm.seismic, vmin=-1, vmax=1)
cb = plt.colorbar(orientation='vertical', fraction=0.046, pad=0.02, label='SLA (m)')
plt.show()

plot_data = raw.ssh.values[0]

fig = plt.figure(figsize=(12, 4))
plt.imshow(np.flipud(plot_data.T), cmap=cm.seismic, vmin=-1, vmax=1)
plt.show()

plot_data = filtered.ssh.values[0]

fig = plt.figure(figsize=(12, 4))
plt.imshow(np.flipud(plot_data.T), cmap=cm.seismic, vmin=-1, vmax=1)
plt.show()
```
